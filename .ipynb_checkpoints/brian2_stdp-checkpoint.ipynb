```python
from brian2 import *

# Parameters
N = 30  # total neurons
duration = 600 * ms
repeats = 20
tau = 10*ms
v_thresh = 1.0
v_reset = 0.0
A_plus = 0.01
A_minus = -0.012
w_max = 0.05

# Neuron model (LIF)
eqs = '''
dv/dt = (-v) / tau : 1
'''

# Neuron groups: one shared group, use input to control sequence
G = NeuronGroup(N, eqs, threshold='v>v_thresh', reset='v=0', method='exact')
G.v = 0

# External inputs: spike generators for A, B, C
indices_A = [0, 1, 2]
indices_B = [10, 11, 12]
indices_C = [20, 21, 22]

spike_times = []

# Create repeated sequences: A at t, B at t+50 ms, C at t+100 ms
for r in range(repeats):
    t_base = r * 200 * ms
    spike_times += [(i, t_base) for i in indices_A]
    spike_times += [(i, t_base + 50*ms) for i in indices_B]
    spike_times += [(i, t_base + 100*ms) for i in indices_C]

#input_group = SpikeGeneratorGroup(N, *zip(*spike_times)*ms)

import numpy as np

spike_indices, spike_times_ms = zip(*spike_times)
spike_indices = list(spike_indices)
spike_times_ms = np.array(spike_times_ms) * ms


# Plot spikes
import matplotlib.pyplot as plt
#plt.figure(figsize=(10, 4))
#plt.plot(spike_mon.t/ms, spike_mon.i, '.k')
#plt.xlabel('Time (ms)')
#plt.ylabel('Neuron index')
#plt.title('Sequence Learning via STDP')
#plt.show()
```

```python
# === Define all groups, synapses, monitors ===

#L1
# Neuron groups: one shared group, use input to control sequence
G = NeuronGroup(N, eqs, threshold='v>v_thresh', reset='v=0', method='exact')
G.v = 0

input_group = SpikeGeneratorGroup(N, spike_indices, spike_times_ms)


# Connect input to neurons (one-to-one)
input_syn = Synapses(input_group, G, on_pre='v_post += 0.6')
input_syn.connect(j='i')

# L1 ecurrent synapses with STDP
syn = Synapses(G, G,
    '''
    w : 1
    dpre/dt = -pre / tau : 1 (event-driven)
    dpost/dt = -post / tau : 1 (event-driven)
    ''',
    on_pre='''
    v_post += w
    pre += 1.
    w = clip(w + A_plus * post, 0, w_max)
    ''',
    on_post='''
    post += 1.
    w = clip(w + A_minus * pre, 0, w_max)
    ''')
syn.connect(p=0.1)
syn.w = 'rand() * 0.01'

# Monitors
spike_mon = SpikeMonitor(G)
state_mon = StateMonitor(G, 'v', record=True)

# === L2 Neurons: Sequence Detectors ===
#L2 = NeuronGroup(3, eqs, threshold='v>v_thresh', reset='v=0', method='exact')
L2 = NeuronGroup(3, eqs, threshold='v > 0.6', reset='v = 0', method='exact')

L2.v = 0

# === Synapses from L1 to L2 (Hebbian learning) ===
L1_L2 = Synapses(G, L2,
    '''
    w : 1
    dpre/dt = -pre / tau : 1 (event-driven)
    dpost/dt = -post / tau : 1 (event-driven)
    ''',
    on_pre='''
    v_post += w
    pre += 1.
    w = clip(w + A_plus * post, 0, w_max)
    ''',
    on_post='''
    post += 1.
    w = clip(w + A_minus * pre, 0, w_max)
    '''
)

L1_L2.connect(p=1.0)
L1_L2.w = '0.05 + rand() * 0.02'

#L1_L2.connect(p=0.3)  # You can customize which L1 neurons connect
#L1_L2.w = 'rand() * 0.01'

spike_mon_L2 = SpikeMonitor(L2)
state_mon_L2 = StateMonitor(L2, 'v', record=True)

# === Feedback from L2 to L1 ===
L2_L1 = Synapses(L2, G, on_pre='v_post += 0.3')  # Tune this weight

# Example: hardcoded prediction mapping
# Assume neuron 0 in L2 predicts B (10–12), neuron 1 predicts C (20–22)
L2_L1.connect(i=0, j=[10, 11, 12])  # L2[0] predicts group B
L2_L1.connect(i=1, j=[20, 21, 22])  # L2[1] predicts group C


# === Create explicit network ===
net = Network()
net.add(G, input_group, input_syn, syn, L2, L1_L2,
        spike_mon, spike_mon_L2,L2_L1)

# === Run the network ===
net.run(repeats * 200 * ms)
```

```python
# Plot L1 and L2 activity
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(2, 1, 1)
plt.plot(spike_mon.t/ms, spike_mon.i, '.k')
plt.title('Layer 1 (Sequence Neurons)')
plt.ylabel('Neuron index')

plt.subplot(2, 1, 2)
plt.plot(spike_mon_L2.t/ms, spike_mon_L2.i, '.r')
plt.title('Layer 2 (Sequence Detectors)')
plt.xlabel('Time (ms)')
plt.ylabel('L2 Neuron index')

plt.tight_layout()
plt.show()
```

```python
# Testing phase
# === Define new test input: only stimulate A neurons once ===
test_indices = [0, 1, 2]  # A neurons
test_times = [0, 0, 0] * ms  # All fire at t = 0 ms

test_input = SpikeGeneratorGroup(N, test_indices, test_times)

# Connect test input to L1
test_input_syn = Synapses(test_input, G, on_pre='v_post += 0.6')
test_input_syn.connect(j='i')
```

```python
# === New NeuronGroups for testing ===
G_test = NeuronGroup(N, eqs, threshold='v>1', reset='v=0', method='exact')
G_test.v = 0

L2_test = NeuronGroup(3, eqs, threshold='v>0.6', reset='v=0', method='exact')
L2_test.v = 0

# === Synapses (reuse trained weights from training phase) ===
# (if you want real carry-over, you'd save weights from G_train & L2_train here)

test_syn = Synapses(G_test, G_test, ...)  # recurrent L1 STDP (optional for recall)
test_L1_L2 = Synapses(G_test, L2_test, ...)  # trained feedforward to L2
test_L2_L1 = Synapses(L2_test, G_test, on_pre='v_post += 0.3')  # feedback

# Example: reconnect L2 → L1 as in training
test_L2_L1.connect(i=0, j=[10,11,12])
test_L2_L1.connect(i=1, j=[20,21,22])

# === Test Input: Only stimulate A neurons ===
test_input = SpikeGeneratorGroup(N, [0,1,2], [0,0,0]*ms)
test_input_syn = Synapses(test_input, G_test, on_pre='v_post += 0.6')
test_input_syn.connect(j='i')

# === Monitors ===
spike_mon_test = SpikeMonitor(G_test)
spike_mon_L2_test = SpikeMonitor(L2_test)

# === Test network ===
test_net = Network()
test_net.add(G_test, L2_test,
             test_input, test_input_syn,
             test_syn, test_L1_L2, test_L2_L1,
             spike_mon_test, spike_mon_L2_test)

test_net.run(100*ms)

```

```python
# Plot test output
plt.figure(figsize=(12, 5))

plt.subplot(2, 1, 1)
plt.plot(spike_mon.t/ms, spike_mon.i, '.k')
plt.title('Layer 1 (Recall Test)')
plt.ylabel('Neuron index')

plt.subplot(2, 1, 2)
plt.plot(spike_mon_L2.t/ms, spike_mon_L2.i, '.r')
plt.title('Layer 2 (Sequence Detectors)')
plt.xlabel('Time (ms)')
plt.ylabel('L2 Neuron index')

plt.tight_layout()
plt.show()
```
